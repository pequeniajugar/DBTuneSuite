```python
import random

# Set seed for reproducibility
random.seed(42)

# Define range and sample size
range_min, range_max = 1, 100000
sample_size = 100

# Select 100 random numbers
selected_numbers = random.sample(range(range_min, range_max + 1), sample_size)

# Define values of n
n_values = [0, 20, 40, 60, 80, 100]

# Generate queries and create separate files for each n value
file_paths = []
for n in n_values:
    file_path = f"queries_no_voj_{n}.txt"
    with open(file_path, "w") as file:
        # Randomly select n numbers for `SELECT id, balance FROM account where id = $;`
        select_balance_ids = random.sample(selected_numbers, n)
        # Remaining for `SELECT * FROM account where id = $;`
        select_all_ids = list(set(selected_numbers) - set(select_balance_ids))

        queries = []
        for id_value in select_balance_ids:
            queries.append(f"SELECT id, balance FROM account where id = {id_value};")
        for id_value in select_all_ids:
            queries.append(f"SELECT * FROM account where id = {id_value};")

        # Shuffle queries to mix both types randomly
        random.shuffle(queries)

        # Write to file
        file.write(f"-- Queries for n = {n}\n")
        file.write("\n".join(queries) + "\n\n")

    file_paths.append(file_path)

# Return all file paths
file_paths
```
